```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 세션 순서형
order = [0.25, 0.5, 0.75, 1.0]
df_all["session"] = pd.Categorical(df_all["session"], categories=order, ordered=True)

def box_with_dots(df, var, title=None, jitter=0.08, dot_size=18, ylim=None):
    """
    session(카테고리)별 박스플롯 + 모든 관측치 점으로 표시
    - df[var]: 수치
    - df['session']: 카테고리(0.25/0.5/0.75/1.0)
    - jitter: 점의 수평흔들림 크기
    - dot_size: 점 크기
    - ylim: (ymin, ymax) 지정 가능
    """
    # 정리
    d = df[["session", var]].dropna().copy()
    if d.empty:
        print(f"[skip] {var}: 데이터가 없습니다.")
        return

    # 카테고리 순서에 맞춰 그룹
    cats = [c for c in order if c in d["session"].cat.categories and (d["session"] == c).any()]
    data_by_cat = [d.loc[d["session"] == c, var].values for c in cats]

    # 그림
    fig, ax = plt.subplots(figsize=(6, 4))
    # 박스플롯
    bp = ax.boxplot(data_by_cat, showfliers=False)

    # 모든 점 찍기(카테고리마다 살짝 좌우로 흩뿌림)
    rng = np.random.default_rng(42)  # 재현가능한 jitter
    for xi, ys in enumerate(data_by_cat, start=1):
        x_jit = rng.uniform(low=xi - jitter, high=xi + jitter, size=len(ys))
        ax.scatter(x_jit, ys, s=dot_size, alpha=0.7)

    # 축/제목
    ax.set_xticks(range(1, len(cats) + 1))
    ax.set_xticklabels([str(c) for c in cats])
    ax.set_xlabel("Threshold (session)")
    ax.set_ylabel(var)
    if title is None:
        title = f"{var} by session (box + all points)"
    ax.set_title(title)
    if ylim is not None:
        ax.set_ylim(*ylim)
    plt.tight_layout()
    plt.show()

# 예시: 활동 비율/가중비율/성과 변수들 몇 개 그리기
vars_to_plot = [
    "chatFrac","classifyFrac","filterFrac","verifyFrac",
    "chatWeight","classifyWeight","filterWeight","verifyWeight",
    "fractionalScore","precision","recall","f1"
]
for v in vars_to_plot:
    if v in df_all.columns:
        box_with_dots(df_all, v)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

order = [0.25, 0.5, 0.75, 1.0]
df_all["session"] = pd.Categorical(df_all["session"], categories=order, ordered=True)

def box_with_dots_and_reg(df, var, title=None, jitter=0.08, dot_size=18, ylim=None, show_equation=True):
    """
    session별 박스플롯 + 모든 관측치 점 + '세션(임계값)'에 대한 전체 회귀 직선
    - 회귀식은 Y ~ session(수치) 1차식
    """
    d = df[["session", var]].dropna().copy()
    if d.empty:
        print(f"[skip] {var}: 데이터가 없습니다.")
        return

    # 세션 카테고리 순서
    cats = [c for c in order if c in d["session"].cat.categories and (d["session"] == c).any()]
    data_by_cat = [d.loc[d["session"] == c, var].values for c in cats]

    # ---- 그림 영역 ----
    fig, ax = plt.subplots(figsize=(6, 4))

    # 1) 박스플롯 (outlier 점은 우리가 별도로 찍으므로 숨김)
    ax.boxplot(data_by_cat, showfliers=False)

    # 2) 원시 점(모든 팀) 찍기: 각 카테고리 x위치는 1..len(cats)
    rng = np.random.default_rng(42)
    for xi, ys in enumerate(data_by_cat, start=1):
        x_jit = rng.uniform(low=xi - jitter, high=xi + jitter, size=len(ys))
        ax.scatter(x_jit, ys, s=dot_size, alpha=0.7)

    # 3) 전체 회귀 직선: session을 실수(0.25/0.5/0.75/1.0)로 두고 Y~X 1차식 적합
    #    박스플롯 좌표(1..4)에 맞추기 위해: pos = 1 + (x - 0.25)/0.25
    x_float = d["session"].astype(float).values
    y = d[var].values
    # numpy로 OLS 1차 회귀
    b1, b0 = np.polyfit(x_float, y, 1)  # y ≈ b0 + b1*x
    x_line = np.linspace(min(order), max(order), 100)
    y_line = b0 + b1 * x_line
    pos_line = 1 + (x_line - min(order)) / (order[1]-order[0])  # 1..4로 매핑
    ax.plot(pos_line, y_line, linewidth=2)

    # (선택) 회귀식/기울기 표기
    if show_equation:
        ax.text(0.02, 0.98,
                f"{var}: y = {b0:.3f} + {b1:.3f}·session",
                transform=ax.transAxes, va="top", ha="left")

    # 축/제목/범위
    ax.set_xticks(range(1, len(cats) + 1))
    ax.set_xticklabels([str(c) for c in cats])
    ax.set_xlabel("Threshold (session)")
    ax.set_ylabel(var)
    if title is None:
        title = f"{var} by session"
    ax.set_title(title)
    if ylim is not None:
        ax.set_ylim(*ylim)
    plt.tight_layout()
    plt.show()

for v in ["chatFrac","chatWeight","classifyFrac","classifyWeight",
          "fractionalScore","precision","recall","f1"]:
    if v in df_all.columns:
        yl = (0,1) if v in ["chatFrac","classifyFrac","filterFrac","verifyFrac",
                            "chatWeight","classifyWeight","filterWeight","verifyWeight",
                            "precision","recall","f1"] else None
        box_with_dots_and_reg(df_all, v, ylim=yl)
# try/except 안에 넣어 회귀계수 p-value 표기
try:
    import statsmodels.api as sm
    X = sm.add_constant(x_float)
    mod = sm.OLS(y, X).fit()
    b0_sm, b1_sm = mod.params
    p_sm = mod.pvalues[1]
    # 텍스트 교체
    ax.text(0.02, 0.98,
            f"{var}: y = {b0_sm:.3f} + {b1_sm:.3f}·session (p={p_sm:.3g})",
            transform=ax.transAxes, va="top", ha="left")
except Exception:
    pass
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats

# 1) 데이터 준비
order = [0.25, 0.5, 0.75, 1.0]
d = df_all[["session", "classifyFrac"]].dropna().copy()
d["session"] = pd.Categorical(d["session"], categories=order, ordered=True)

# 2) 일원분산분석 (session을 범주로)
m = smf.ols("classifyFrac ~ C(session)", data=d).fit()
aov = sm.stats.anova_lm(m, typ=2)   # Type-II ANOVA
print(aov)

# 3) 효과크기 (보고서용)
ss_between = aov.loc["C(session)", "sum_sq"]
ss_within  = aov.loc["Residual",   "sum_sq"]
df_between = aov.loc["C(session)", "df"]
df_within  = aov.loc["Residual",   "df"]
ss_total   = ss_between + ss_within
ms_within  = ss_within / df_within

eta2   = ss_between / ss_total
omega2 = (ss_between - df_between * ms_within) / (ss_total + ms_within)

print(f"\nR²(모형): {m.rsquared:.3f}")
print(f"η²: {eta2:.3f}   ω²: {omega2:.3f}")
print(f"ANOVA p-value: {aov.loc['C(session)', 'PR(>F)']:.4g}")

# (선택) 등분산 가정 점검
lev = stats.levene(*[d.loc[d['session']==s, 'classifyFrac'].values for s in order])
print(f"Levene test for equal variances: p={lev.pvalue:.4g}")

# (선택) 등분산이 깨지면 Welch ANOVA도 참고 (statsmodels 0.13+)
try:
    from statsmodels.stats.oneway import anova_oneway
    welch = anova_oneway([d.loc[d['session']==s, 'classifyFrac'].values for s in order],
                         use_var='unequal', welch_correction=True)
    print("\nWelch ANOVA (unequal variances allowed)")
    print(welch)
except Exception:
    pass
```

```python
import numpy as np
import pandas as pd
from scipy import stats

order = [0.25, 0.5, 0.75, 1.0]
d = df_all[["session", "classifyFrac"]].dropna().copy()
d = d[d["session"].isin(order)]

groups = [d.loc[d["session"]==s, "classifyFrac"].to_numpy() for s in order]
k = len(groups)
ns = [len(g) for g in groups]
N = sum(ns)

lev = stats.levene(*groups, center='median')  # 등분산이면 p>=.05
print(f"Levene test : stat={lev.statistic:.3f}, p={lev.pvalue:.4g}")

F, p = stats.f_oneway(*groups)
df_between = k - 1
df_within  = N - k
print(f"\nOne-way ANOVA for classifyFrac ~ session")
print(f"p = {p:.4g}")

# 4) 효과크기 계산 (수식으로 직접)
# 총평균
overall_mean = np.concatenate(groups).mean()

# 집단간 제곱합 SSB, 집단내 제곱합 SSW
SSB = sum(n * (g.mean() - overall_mean)**2 for g, n in zip(groups, ns))
SSW = sum(((g - g.mean())**2).sum() for g in groups)
SST = SSB + SSW
MSW = SSW / df_within

eta2   = SSB / SST
omega2 = (SSB - df_between * MSW) / (SST + MSW)


print(f"This hypothesis(h1) can be successfully rejected because p-value is lower that 0.1")
```

```python
# 1. 전략 차이 계산(classify<>filter)
df_all['strategy_diff'] = df_all['g_classifyFrac'] - df_all['g_filterFrac']

# 2. 전략이 filter > classify로 전환되는 시점 찾기
def find_switch_point(group):
    group_sorted = group.sort_values('session')
    for i in range(len(group_sorted)):
        if group_sorted.iloc[i]['strategy_diff'] < 0:
            return group_sorted.iloc[i]['session']
    return 1.1  # 전략 전환 안 한 경우

switch_df = df_all.groupby('instanceId').apply(find_switch_point).reset_index()
switch_df.columns = ['instanceId', 'switch_point']

# 3. 최종 성과와 병합
final_score = df_all[df_all['session'] == 1.0][['instanceId', 'g_fractionalScore']]
merged = pd.merge(switch_df, final_score, on='instanceId')

# 확인
merged['switch_point'].value_counts().sort_index()
merged.describe()
```

```python
# 필요한 열만 추출해서 피벗
df_classify = df_all[['instanceId', 'session', 'classifyFrac', 'g_fractionalScore']]

# pivot_table 사용하여 평균으로 통합
df_pivot = df_classify.pivot_table(
    index='instanceId',  # 팀 ID
    columns='session',   # 세션
    values='classifyFrac',
    aggfunc='mean'
).reset_index()

# 열 이름 정리
df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    0.25: 'classifyFrac_0.25',
    0.5: 'classifyFrac_0.5',
    0.75: 'classifyFrac_0.75',
    1.0: 'classifyFrac_1.0'
})


def detect_turning_point(row):
    d1 = row['classifyFrac_0.5'] - row['classifyFrac_0.25']
    d2 = row['classifyFrac_0.75'] - row['classifyFrac_0.5']
    d3 = row['classifyFrac_1.0'] - row['classifyFrac_0.75']

    if d1 * d2 < 0:
        return 0.5
    elif d2 * d3 < 0:
        return 0.75
    else:
        return 1.1  # No clear turning point

df_pivot['turning_point'] = df_pivot.apply(detect_turning_point, axis=1)

# nominal 데이터에서 성과 가져오기
df_score = df_nominal[['instanceId', 'g_fractionalScore']].drop_duplicates()

# 결합
df_result = df_pivot.merge(df_score, on='instanceId', how='left')


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_result, x='turning_point', y='g_fractionalScore')
plt.title('Performance by First ClassifyFrac Turning Point')
plt.ylabel('Final Team Performance (g_fractionalScore)')
plt.xlabel('Turning Point (Session)')
plt.grid(True)
plt.show()

# ANOVA
group_05 = df_result[df_result['turning_point'] == 0.5]['g_fractionalScore']
group_075 = df_result[df_result['turning_point'] == 0.75]['g_fractionalScore']
group_none = df_result[df_result['turning_point'] == 1.1]['g_fractionalScore']

f_stat, p_val = f_oneway(group_05, group_075, group_none)
print(f"\nANOVA result\nF-statistic: {f_stat:.4f}, p-value: {p_val:.4f}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 필요한 열 추출
df_slim = df_all[['instanceId', 'session', 'g_effortEntropy', 'g_fractionalScore']]

# pivot: 세션별 분리
pivot_entropy = df_slim.pivot_table(index='instanceId', columns='session', values='g_effortEntropy')
pivot_score = df_slim.pivot_table(index='instanceId', columns='session', values='g_fractionalScore')

# 이전 세션 성과가 다음 세션 균등성에 영향을 주는지 확인
df_feedback = pd.DataFrame({
    'score_025': pivot_score[0.25],
    'entropy_050': pivot_entropy[0.50],
    'score_050': pivot_score[0.50],
    'entropy_075': pivot_entropy[0.75],
    'score_075': pivot_score[0.75],
    'entropy_100': pivot_entropy[1.00]
})

def feedback_regression(x, y, x_label, y_label):
    X = x.values.reshape(-1, 1)
    Y = y.values

    model = LinearRegression()
    model.fit(X, Y)
    y_pred = model.predict(X)
    r2 = r2_score(Y, y_pred)
    b0 = model.intercept_
    b1 = model.coef_[0]

    # 시각화
    plt.figure(figsize=(7, 5))
    sns.scatterplot(x=x, y=y)
    plt.plot(x, y_pred, color='red', linestyle='--')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f'{x_label} → {y_label}\nR²={r2:.4f}, B1={b1:.4f}, B0={b0:.4f}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print(f"{x_label} → {y_label}")
    print(f"Intercept (B0): {b0:.4f}")
    print(f"Slope     (B1): {b1:.4f}")
    print(f"R² Score      : {r2:.4f}")
    print()
feedback_regression(df_feedback['score_025'], df_feedback['entropy_050'],
                    'Score at 0.25', 'Entropy at 0.50')

feedback_regression(df_feedback['score_050'], df_feedback['entropy_075'],
                    'Score at 0.50', 'Entropy at 0.75')

feedback_regression(df_feedback['score_075'], df_feedback['entropy_100'],
                    'Score at 0.75', 'Entropy at 1.00')
```
